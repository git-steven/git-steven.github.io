## 📊 2 — Jupyter Notebook Code (Python + matplotlib)

Below is a notebook you *can run*, generating the 3D “prop” pyramid and the symmetric horizontal funnel with realistic slopes. It uses matplotlib’s 3D support and basic bar plotting to approximate your diagram.

You can customize the palette using values from your SVG (just replace the `colors` list).

````python

```python
# di_kw_funnel.ipynb
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ------- Settings ------
levels = ["Data", "Information", "Knowledge", "Wisdom"]
values = [1.0, 0.75, 0.5, 0.25]  # relative “widths”
colors = ["#6da7d1", "#8dbb6f", "#f2c75c", "#e79a3e"]  # placeholder palette

# --- 3D Pyramid side view (sloping) ---
fig = plt.figure(figsize=(12, 6))
ax3 = fig.add_subplot(121, projection='3d')
z_height = [0, 1, 2, 3]
x_width = [1.0 - v*0.5 for v in values]
y_width = values

for i, lvl in enumerate(levels):
    xi = x_width[i]
    yi = y_width[i]
    zi = z_height[i]
    # draw trapezoidal “slice”
    ax3.bar3d(xi/2, yi/2, zi, xi, yi, 0.5, color=colors[i], alpha=0.8)
    ax3.text(xi/2 + xi/3, yi/2 + yi/3, zi+0.25, lvl, color='k')

ax3.set_title("DIKW Pyramid (3D prop view)")
ax3.set_xlabel("Width")
ax3.set_ylabel("Depth")
ax3.set_zlabel("Level")
ax3.set_xticks([])
ax3.set_yticks([])
ax3.set_zticks([])

# --- Horizontal Funnel View ---
ax2 = fig.add_subplot(122)
positions = np.arange(len(levels))
ax2.barh(positions, values[::-1], color=colors[::-1], height=0.6)
ax2.set_yticks(positions)
ax2.set_yticklabels(levels[::-1])
ax2.set_title("Horizontal Funnel (signal narrows toward wisdom)")
ax2.set_xlabel("Relative focus")
ax2.set_xlim(0, 1.1)
plt.tight_layout()
plt.show()
````
